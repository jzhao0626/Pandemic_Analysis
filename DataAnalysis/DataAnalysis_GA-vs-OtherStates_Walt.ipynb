```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
# Import the dataset
csvpath = '../DataCleaning/Output/StatesCases.csv'
final_df = pd.read_csv(csvpath)
final_df
```

```python
# Plot the cumulative cases of each state in a bar graph
final_df.plot(kind='bar', x='State', y='Cumulative Cases', figsize=(16,8), rot = 0, title='Cumulative Cases of Each State')
plt.ylabel('Cumulative Cases')
```

```python
# Plot the cumulative death of each state in a bar graph
final_df.plot(kind='bar', x='State', y='Cumulative Deaths', figsize=(16,8), rot = 0, title='Cumulative Deaths of Each State', color='r')
plt.ylabel('Cumulative Death')
```

```python
# Plot the cumulative cases of each state in a bar graph sorted by the number of cases
sorted_cases_df = final_df.sort_values('Cumulative Cases', ascending=False)
sorted_cases_df.plot(kind='bar', x='State', y='Cumulative Cases', figsize=(20,8), rot = 0, title='Cumulative Cases of Each State')
plt.ylabel('Cumulative Cases')
plt.savefig('Figures/Cumulative_Cases_of_Each_State.png')
```

```python
# Plot the cumulative death of each state in a bar graph sorted by the number of death
sorted_deaths_df = final_df.sort_values('Cumulative Deaths', ascending=False)
sorted_deaths_df.plot(kind='bar', x='State', y='Cumulative Deaths', figsize=(20,8), rot = 0, title='Cumulative Deaths of Each State', color='r')
plt.ylabel('Cumulative Death')
plt.savefig('Figures/Cumulative_Deaths_of_Each_State.png')
```

```python
# Plot the cumulative cases of top 10 state in a bar graph sorted by the number of cases and GA highlighted
top_cases_df = sorted_cases_df.iloc[0:10]
plt.figure(figsize=(16,4))
top_case_graph = plt.bar(top_cases_df['State'], top_cases_df['Cumulative Cases'])
top_case_graph[5].set_color('r')
plt.title('States of Most Cumulative Cases')
plt.xlabel('State')
plt.ylabel('Cumulative Cases')
plt.savefig('Figures/Cumulative_Cases_of_Top_State.png')
```

```python
# Plot the cumulative death of top 10 state in a bar graph sorted by the number of death and GA highlighted
top_deaths_df = sorted_deaths_df.iloc[0:10]
plt.figure(figsize=(16,4))
top_death_graph = plt.bar(top_deaths_df['State'], top_deaths_df['Cumulative Deaths'], color='r')
top_death_graph[8].set_color('#1f77b4')
plt.title('States of Most Cumulative Deaths')
plt.xlabel('State')
plt.ylabel('Cumulative Deaths')
plt.savefig('Figures/Cumulative_Deaths_of_Top_State.png')
```
