```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
csvpath1 = '../DataCleaning/Resources/TOP_Counties_GA_cases_hosp_deaths_by_age_group.csv'
age_df = pd.read_csv(csvpath1)
age_df
```

```python
cases_age_df = age_df.pivot(index='county name', columns='age group', values = 'cases')
cases_age_df
cases_age_df.plot.bar(figsize=(16,8),rot=0,xlabel = 'County', ylabel='Cases', title='Cases of Top County')
plt.savefig('Figures/Cases_of_Top_County.png')
```

```python
deaths_age_df = age_df.pivot(index='county name', columns='age group', values = 'deaths')
deaths_age_df
deaths_age_df.plot.bar(figsize=(16,8),rot=0,xlabel = 'County', ylabel='Deaths', title='Deaths of Top County')
plt.savefig('Figures/Deaths_of_Top_County.png')
```

```python
hospitalization_age_df = age_df.pivot(index='county name', columns='age group', values = 'hospitalization')
hospitalization_age_df
hospitalization_age_df.plot.bar(figsize=(16,8),rot=0,xlabel = 'County', ylabel='Hospitalization', title='Hospitalization of Top County')
plt.savefig('Figures/Hospitalization_of_Top_County.png')
```
