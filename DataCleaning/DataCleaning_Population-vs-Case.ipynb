```python
# Dependencies
import pandas as pd
```

```python
csvpath1 = 'Resources/Average_Household_Size_and_Population_Density_-_County.csv'
PopDen_df = pd.read_csv(csvpath1)
PopDen_df
```

```python
csvpath2 = 'Resources/GA_county_cases.csv'
Case_df = pd.read_csv(csvpath2)
Case_df
```

```python
filtered_PopDen_df = PopDen_df[['State','NAME','B01001_calc_PopDensity']]
cleaned_PopDen_df = filtered_PopDen_df.rename(columns={'NAME':'County', 'B01001_calc_PopDensity':'PopulationDensity'})
cleaned_GA_df = cleaned_PopDen_df.loc[cleaned_PopDen_df['State'] == 'Georgia'][['County','PopulationDensity']]
cleaned_GA_df
```

```python
filtered_Case_df = Case_df[['county_name', 'cases','hospitalization','deaths','case rate','death rate']]
cleaned_Case_df = filtered_Case_df.rename(columns={'county_name':'County',
                                 'cases':'Cases',
                                 'hospitalization':'Hospitalization',
                                 'deaths':'Deaths',
                                 'case rate':'Cases/100,000Population',
                                 'death rate':'Deaths/100,000Population'})
cleaned_Case_df
```

```python
merge_df = cleaned_Case_df.merge(cleaned_GA_df, on='County')
final_df = merge_df.set_index('County')
```

```python
final_df.to_csv('Output/PopulationDensity.csv')
```
